```python
# Escribe el formato en webm
nombre_archivo='numero'

with open(f'{nombre_archivo}.webm', 'wb') as f:
    f.write(recorder.audio.value)

# Convierte en webm -> wav
!ffmpeg -i numero.webm -ac 1 -f wav ./wav/numero.wav -y -hide_banner -loglevel panic

audio_paths = [f"./wav/{nombre_archivo}.wav"]

# Obtenemos la frase del usuario
transcriptions_es = model.transcribe(audio_paths)[0]
# Obtener la frase
transcriptions_es['transcription']
```
